# Define functions

```python
add_text = 'std_4_Hetero'
```

```python
K = 2
nseq = [200]*K
means = [-3,3]
stds = [4]*2
scale = 0.1

test_mv = [0, 3]
# sample data group different groups
```

```python
def w1(x):
    v = 9
    mm = -3
    vv = 16
    return np.exp(-0.5/v *(x**2) + 0.5/vv*(x-mm)**2)

def w2(x):
    v = 9
    mm = 3
    vv = 16
    return np.exp(-0.5/v *(x**2) + 0.5/vv*(x-mm)**2)
    return 
```

# Computation

```python
def scale_fun(x):
    return 0.03*( np.abs(x - 5)**0.5 * np.abs(x + 5)**0.5  + 1) 
```

```python
# pretrain a GP model
np.random.seed(324)
low, high, sample_size = -15, 15, 100
Xtr = np.random.uniform(low, high, sample_size)
Ytr = generate_response_h(Xtr, scale_fun)

kernel = GPy.kern.RBF(input_dim=1)
model = GPy.models.GPRegression(Xtr.reshape(-1,1), Ytr.reshape(-1,1), kernel)

# optimize the model parameters
model.optimize()
f = model.predict
```

```python
# prepare calibration data and test data
Xcal, Ycal, Xtest, Ytest = generate_data_h(nseq, means, stds, scale_fun, test_mv)
```

```python
# get prediction band from -10 to 10
ns = 16
xseq = np.linspace(-11, 11, ns).reshape(-1)
```

```python
CI = np.zeros((4,ns,2))
qn = scipy.stats.norm.ppf(0.95)
```

```python
for i in range(ns):
    x = xseq[i]
    PIlist = dwcp_interval(x, f,  Xcal, Ycal, [w1, w2], 0.1)
    for k in range(3):
        CI[k, i,:] = PIlist[k,:]
    CI[3,i,0] = sigmoid(x) - qn*scale_fun(x)
    CI[3,i,1] = sigmoid(x) + qn*scale_fun(x)
```

```python
x = np.linspace(-10, 10, 50).reshape(-1)
y = sigmoid(x)

# Example array of confidence intervals
ms = 8
mso = 6
lwd = 2
colors = ['blue', 'red', 'purple']

fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y, color='black', linestyle = 'dashed', 
         linewidth=2, alpha = 0.8, label='E(Y|X)')

plt.fill_between(xseq, CI[3,:,0], CI[3,:,1], color='black', alpha=0.2, label='Oracle 90% CI')

handles, labels = plt.gca().get_legend_handles_labels()

k = 0
for i, ci in enumerate(CI[k,:,:]):
    ax.plot(xseq[i], CI[k,i,0], 'o', markersize=ms, fillstyle='full', color = colors[k], alpha = 0.8)
    ax.plot(xseq[i], CI[k,i,1], 'o', markersize=ms, fillstyle='full', color = colors[k], alpha = 0.8)
    if CI[k,i,1] == np.inf:
        ax.plot(np.array([xseq[i], xseq[i]]), [-10,10], 
            'k-', linestyle=':', linewidth= lwd, color = colors[k]) 
    else:
        ax.plot(np.array([xseq[i], xseq[i]]), CI[k,i,:], 
            'k-', linewidth= lwd, color = colors[k])
    

    

# create manual symbols for legend  
# line = Line2D([0], [0], label='manual line', color='blue')
lines = Line2D([0], [0], color='blue', linestyle=':', linewidth=lwd,label = 'WCP-infinity')

point = Line2D([0], [0], label='WCP-finite', marker='o', markersize=ms, color = colors[k], alpha = 0.8,
         linestyle='solid')

# add manual symbols to auto legend
handles.extend([lines, point])

    
        
plt.xlabel('Feature', fontsize=20)
plt.ylabel('Response', fontsize=20)
plt.title('Prediction band by WCP on group 1', fontsize=24, y = 1.02)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.3)
led = ax.legend(handles=handles, fontsize = 20, borderaxespad=0.15)
led.get_frame().set_alpha(None)
led.get_frame().set_facecolor('white')
# ax.legend(['E(Y|X)', 'Oracle CIs',  'WCP1-info', 'WCP1-non-info' ], fontsize=12)


# Set plot limits
plt.xlim([-10, 10])
plt.ylim([-0.5, 1.5])
plt.savefig('/Users/mufangying/Desktop/DWCP/figures/'+ add_text+'WCP_1.pdf')
```

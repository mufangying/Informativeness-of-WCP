# Computation

```python
Xcal, Ycal, Xtest, Ytest = generate_data(nseq, means, stds, scale, test_mv)
# get prediction band from -10 to 10
ns = 16
xseq = np.linspace(-11, 11, ns).reshape(-1)
```

```python
CI = np.zeros((6,ns,2))
qn = scipy.stats.norm.ppf(0.95)
```

```python
for i in range(ns):
    x = xseq[i]
    PIlist = get_intervals(x, f,  Xcal, Ycal, w, w_pool, nseq, means, stds, test_mv, 0.1)
    for k in range(5):
        CI[k, i,:] = PIlist[k,:]
    CI[5,i,0] = sigmoid(x) - qn*scale
    CI[5,i,1] = sigmoid(x) + qn*scale
```

```python
x = np.linspace(-10, 10, 50).reshape(-1)
y = sigmoid(x)

# Example array of confidence intervals
ms = 8
mso = 6
lwd = 2
colors = ['blue', 'red', 'purple']

fig, ax = plt.subplots(figsize=(12, 8))

ax.plot(x, y, color='black', linestyle = 'dashed', 
         linewidth= 4, alpha = 0.8, label='E(Y|X)')

plt.fill_between(xseq, CI[5,:,0], CI[5,:,1], color='black', alpha=0.2, label='Oracle 90% CI')

handles, labels = plt.gca().get_legend_handles_labels()

k = 0
for i, ci in enumerate(CI[k,:,:]):
    ax.plot(xseq[i], CI[k,i,0], 'o', markersize=ms, fillstyle='full', color = colors[k], alpha = 0.8)
    ax.plot(xseq[i], CI[k,i,1], 'o', markersize=ms, fillstyle='full', color = colors[k], alpha = 0.8)
    if CI[k,i,1] == np.inf:
        ax.plot(np.array([xseq[i], xseq[i]]), [-10,10], 
            'k-', linestyle=':', linewidth= lwd, color = colors[k]) 
    else:
        ax.plot(np.array([xseq[i], xseq[i]]), CI[k,i,:], 
            'k-', linewidth= lwd, color = colors[k])
    

    

# create manual symbols for legend  
# line = Line2D([0], [0], label='manual line', color='blue')
lines = Line2D([0], [0], color='blue', linestyle=':', linewidth=lwd,label = 'Uninformative PI')

point = Line2D([0], [0], label='Informative PI', marker='o', markersize=ms, color = colors[k], alpha = 0.8,
         linestyle='solid')

# add manual symbols to auto legend
handles.extend([lines, point])

    
        
plt.xlabel('Feature', fontsize=30)
plt.ylabel('Response', fontsize=30)
plt.title('Prediction band by WCP on Group 1', fontsize=35, y = 1.02)
# plt.grid(True, linestyle='--', alpha=0.3)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, linestyle='--', alpha=0.3)
led = ax.legend(handles=handles, fontsize = 25, borderaxespad=0.15)
led.get_frame().set_alpha(None)
led.get_frame().set_facecolor('white')
# ax.legend(['E(Y|X)', 'Oracle CIs',  'WCP1-info', 'WCP1-non-info' ], fontsize=12)


# Set plot limits
plt.xlim([-10, 10])
plt.ylim([-0.5, 1.5])
plt.savefig('/Users/mufangying/Desktop/ICP/figures/'+ add_text+'WCP_1.pdf')
```

# Compute the coverage in 5000 replications

```python
R = 5000
covs, lens, counts = replications(R, f, w, w_pool,  nseq, means, stds, test_mv, scale)
# WCP then merged

print('Marginal coverage:')
print(np.round(np.mean(covs, axis = 0), decimals = 3))
print('')

print('Probabiilty of getting finite prediction intervals')
print(np.round(np.mean(counts, axis = 0), decimals = 3))
print('')


print('Informative coverage probability:')
print( np.round( np.sum(counts*covs, axis = 0) /np.sum(counts, axis = 0), decimals = 3)  )
print('')



avg_finite_len = []
for k in range(5):
    lenk = lens[:,k]
    lenk = lenk[~np.isinf(lenk)]
    if len(lenk) == 0:
        avg_finite_len.append(  np.inf )
    else:
        avg_finite_len.append( np.round( np.mean(lenk), decimals=3) )
print('Lengths of informative intervals:')
print(avg_finite_len )
print('')
```
